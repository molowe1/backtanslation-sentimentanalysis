```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#open the tweets
with open("mandarin.txt", "r", encoding="utf-8") as files:
    lines = files.readlines()


for line in lines:
    print(line.strip())
```

```python
#Create an empty list
mandarin_list = []

#Iterate over the txt file
for line in lines:
    mandarin_list.append(line)
    
#get into data frame
mandarin_df = pd.DataFrame({"mandarin_comments":mandarin_list})
```

```python
import re
# Function to remove numbers and punctuation using regular expressions
def remove_numbers_and_punctuation(text):
    # Replace all numbers and punctuation with an empty string
    cleaned_text = re.sub(r'[0-9!@#$%^&*(),.?":{}|<>]', '', text)
    return cleaned_text

# Apply the function to the 'Text' column in the DataFrame
mandarin_df['Cleaned_Text'] = mandarin_df["mandarin_comments"].apply(remove_numbers_and_punctuation)
```

```python
#translate to english
from googletrans import Translator
import time

# Function to translate text from mandarin to English with a retry mechanism
def translate_to_english(text, max_retry=5, sleep_duration=1.0):
    retry_count = 0
    translator = Translator()
    while retry_count < max_retry:
        try:
            translated_text = translator.translate(text, src='zh-cn', dest='en').text
            return translated_text
        except Exception as e:
            print(f"Translation failed. Retrying... ({retry_count + 1}/{max_retry})")
            retry_count += 1
            time.sleep(sleep_duration)
    print("Translation failed after maximum retries.")
    return None

# Apply the translation function to the 'Mandarin_Text' column in the DataFrame
mandarin_df['English_Text'] = mandarin_df['Cleaned_Text'].apply(translate_to_english)

mandarin_df.drop(columns=["mandarin_comments", "Cleaned_Text"], inplace=True)
mandarin_df
```

```python
# Download the stopwords for English from NLTK
import nltk
nltk.download('stopwords')
nltk.download('punkt')

# to remove english stopwords
!pip install textblob
from textblob import TextBlob
from nltk.corpus import stopwords # get stopwords from NLTK library
from nltk.tokenize import word_tokenize # to create word tokens
from nltk.stem import WordNetLemmatizer # to reduce words to orginal form
from nltk.corpus import words # Get all words in english language
```

```python
# Function to remove English stopwords from text
def remove_english_stopwords(text):
    if pd.notna(text):  # Check if the value is not None
        stop_words = set(stopwords.words('english'))
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
        return ' '.join(filtered_text)
    return text  # Return the original text if it's None

# Apply the function to the 'Text' column in the DataFrame
mandarin_df['English_Text'] = mandarin_df['English_Text'].apply(remove_english_stopwords)
```

```python
#sentiment analysis
# Function to perform sentiment analysis using TextBlob
def get_sentiment(text):
    if text is not None:  # Check if the value is not None
        analysis = TextBlob(text)
        polarity = analysis.sentiment.polarity
        if polarity > 0:
            return 'Positive'
        elif polarity < 0:
            return 'Negative'
        else:
            return 'Neutral'
    return 'Neutral'  # Return 'Neutral' for None values


# Apply the function to the 'Text' column in the DataFrame
mandarin_df['Sentiment'] = mandarin_df['English_Text'].apply(get_sentiment)
mandarin_df
```

```python
# See quick results of the Sentiment Analysis
mandarinsentiment_stored = mandarin_df['Sentiment'].value_counts()
mandarin_df['Sentiment'].value_counts()
```

```python
#performing back translation and sentiment analysis

#drop the sentiment column in the mandarin_df dataframe
mandarin_df.drop(columns=["Sentiment"], inplace=True)
```

```python
#translate to mandarin
# Function to translate text from english to mandarin with a retry mechanism
def translate_to_mandarin(text, max_retry=5, sleep_duration=1.0):
    retry_count = 0
    translator = Translator()
    while retry_count < max_retry:
        try:
            translated_text = translator.translate(text, src='en', dest='zh-cn').text
            return translated_text
        except Exception as e:
            print(f"Translation failed. Retrying... ({retry_count + 1}/{max_retry})")
            retry_count += 1
            time.sleep(sleep_duration)
    print("Translation failed after maximum retries.")
    return None
# Apply the translation function to the 'Mandarin_Tweet' column in the DataFrame
mandarin_df['Mandarin_Translation'] = mandarin_df['English_Text'].apply(translate_to_mandarin)

# drop the DataFrame with English translations
mandarin_df.drop(columns=["English_Text"], inplace=True)
mandarin_df
```

```python
#peforming sentiment analysis
!pip install jieba
!pip install snownlp

import jieba
import snownlp
from snownlp import SnowNLP
```

```python
# Function to perform sentiment analysis on a single tweet
def analyze_sentiment(text):
    if text is None:
        return None
    words = jieba.lcut(text)
    segmented_text = ' '.join(words)
    s = SnowNLP(segmented_text)
    sentiment_score = s.sentiments
    return sentiment_score

# Perform sentiment analysis on each tweet in the DataFrame
texts = mandarin_df['Mandarin_Translation'].tolist()
sentiment_scores = [analyze_sentiment(tweet) for tweet in texts]

# Add sentiment scores to the DataFrame
mandarin_df['Sentiment_Score'] = sentiment_scores

# Interpret the sentiment scores and add sentiment labels to the DataFrame
def interpret_sentiment(score):
    if score is None:
        return None
    if score >= 0.6:
        return "Positive"
    elif score >= 0.4:
        return "Neutral"
    else:
        return "Negative"

mandarin_df['Sentiment_Label'] = mandarin_df['Sentiment_Score'].apply(interpret_sentiment)

# Print the DataFrame with sentiment analysis results
print(mandarin_df[['Mandarin_Translation', 'Sentiment_Label', 'Sentiment_Score']])
```
